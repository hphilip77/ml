Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
ecom.columns
```

```python
y = ecom['Yearly Amount Spent']
```

```python
X = ecom[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#creating a variable for linear regression
customer = LinearRegression()
```

```python
customer.fit(X_train, y_train)
```

```python
print(customer.intercept_)
```

```python
customer.coef_
```

```python
X_train.columns
```

```python
#Creating another variables for Coefficient
customer_df = pd.DataFrame(customer.coef_,X.columns,columns=['Coefficients'])
```

```python
customer_df
```

PREDICTION

```python
predictions = customer.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.histplot(y_test-predictions)
```

Evaluating the Model

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
